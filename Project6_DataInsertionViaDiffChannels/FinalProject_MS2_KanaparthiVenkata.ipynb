### 1. Replace the Headers

```python
names = []
with open('california_names.txt', 'r') as f:
    for line in f:
        f.readline()
        var = line.split(":")[0]
        names.append(var)
names
```

```python
df = pd.read_csv("california.csv", names = names)
df.head(5)
```

### 2. Create a data set with required columns
